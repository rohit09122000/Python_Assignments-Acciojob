# Q15 - Input: Circle calculations

Ask the user to input the radius of a circle (float). Compute and print:

```python
# Answer here
# Hint: use input(), then convert to int() or float() as needed.
radius = float(input("Enter radius: "))

circumference = 2 * 3.14159 * radius
area = 3.14159 * radius ** 2

print("Circumference:", circumference)
print("Area:", area)

```

# Q16 - Input: Convert total seconds to hours, minutes, seconds

Ask the user to input a total number of seconds (integer). Convert it to hours, minutes, and seconds and print as `H:M:S`.

```python
# Answer here
# Hint: use input(), then convert to int() or float() as needed.
total_seconds = int(input("Enter total seconds: "))

hours = total_seconds // 3600
minutes = (total_seconds % 3600) // 60
seconds = total_seconds % 60

print(f"{hours}:{minutes}:{seconds}")
```

# Q17 - Input: Simple interest calculation

Ask user for `principal`, `rate` (annual percent), and `time` (years). Compute simple interest as `(principal * rate * time) / 100` and print `Interest: <result>`.

```python
# Answer here
principal = float(input("Enter principal amount: "))
rate = float(input("Enter rate of interest: "))
time = float(input("Enter time (years): "))

interest = (principal * rate * time) / 100
print("Interest:", interest)

```
